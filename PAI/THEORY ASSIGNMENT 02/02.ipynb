```python
import numpy as np
```

```python
datalist = []
```

```python
with open("sensor_data.csv") as file:
    for line in file:
        row = line.strip().split(",")
        datalist.append(row)
```

```python
data = np.array(datalist , dtype=float)
```

```python
data[data == -999] = np.nan
data[data < 0] = np.nan
data[data > 100] = np.nan
```

```python
sensorMeans = np.nanmean(data,axis=0)
```

```python
hourMedian = np.nanmedian(data, axis=1)
```

```python
nanCounts = np.isnan(data).sum(axis=0)
```

```python
worstSensorIndex = np.argmax(nanCounts)
```

```python
globalMin = np.nanmin(data)
globalMax = np.nanmax(data)
```

```python
normalized = (data-globalMin) / (globalMax-globalMin)
```

```python
np.savetxt("sensor_data_normalized.csv", normalized,delimiter=",")
```
